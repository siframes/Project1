```python
import numpy as np
from matplotlib import pyplot as plt

ys = 13 + np.random.randn(24)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys,  12,200, where=(ys >  14,300  ), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
